```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
# input image dimensions
img_rows, img_cols = 96, 96
input_shape = (img_rows, img_cols, 3)

Input = np.load('Data/selected_data.npz')

data = Input["data"]
label = Input["label"]

print (data.shape)
print (label.shape)
plt.imshow(data[0,...])
```

```python
def rgb0(rgb):
    r, g, b = rgb[:, :, 0], rgb[:, :, 1], rgb[:, :, 2]
    return r
```

```python
newdata = []
for i in range(data.shape[0]):
    temp = rgb0(data[i])
    newdata.append(temp)

newdata = np.array(newdata)
newdata = newdata.reshape((1270, 96, 96, 1))
print(newdata.shape)
```

```python
np.savez('Data/singleChannelData.npz', data = newdata, label = label)
```
